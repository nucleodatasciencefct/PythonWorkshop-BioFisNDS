#### OpenCV is a cross-platform and open source computer vision library developed by Intel and later supported by Willow Garage, written mainly in C/C++.

In this section, we will learn how to display, convert, draw and make operations on images using OpenCV. This are one of the basic processes in terms of image processing.

#### Read/Write images

We will use matplotlib as a support to display the image, which can be done with the command plt.imshow.

First, we need to read the image before displaying. To do this we use the cv2.imread function. The first argument of the function is the directory of the image. We can also specify the type of input color-space we like, which is described by the second argument, a flag which specifies the way image should be read. <br>

- cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag. - 1
- cv2.IMREAD_GRAYSCALE : Loads image in grayscale mode - 0
- cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel - (-1).

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('media/CPU.png', -1)
plt.imshow(image)
```

This can also be done with the cv2.imshow extension. This function will display an image in a window. The window automatically fits to the image size.

First argument is a window name which is a string. Second argument is our image. You can create as many windows as you wish, but with different window names.

**cv2.waitKey()** is a keyboard binding function. Its argument is the time in milliseconds. The function waits for specified milliseconds for any keyboard event. If you press any key in that time, the program continues. If 0 is passed, it waits indefinitely for a key stroke. It can also be set to detect specific key strokes like, if key a is pressed etc which we will discuss below.

**cv2.destroyAllWindows()** simply destroys all the windows we created. If you want to destroy any specific window, use the function cv2.destroyWindow() where you pass the exact window name as the argument.

```python
cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

To write an image, we need to use the function **cv2.imwrite()**.

First argument is the file name, second argument is the image you want to save.

```python
cv2.imwrite('FirstWrite.png',image)
```

#### Convertion

In this section, we will learn how to convert from one color-space to another. We will focus on Gray and HSV transition.


For BGR  Gray conversion we use the flags cv.COLOR_BGR2GRAY. Similarly for BGR  HSV, we use the flag cv.COLOR_BGR2HSV. This  flags are the second argunment of the function cvt.Color, in which the first argument must be the variable name of the original image.

Note that, if we indeed display with the function plt.imshow, we also need to specify and confirm, as a second parameter (cmap) the scale tonality, since the library has its own color-space regulator.

```python
image_gray_converted = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray_converted, cmap='gray')
```

```python
image_gray_converted = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(image_gray_converted)
```

#### Drawing

We can also draw any form in our images. First, we make a copy of our image with image.copy() method. If we draw a circle, we use the cv2.circle method, and you need its center coordinates and radius. 

```python
image_new = image.copy()
cv2.circle(image_new, (500, 500), 7, color = (255, 215, 0), thickness = 100, )
plt.imshow(image_new);
```

We can also use cv2.rectangle to draw a rectangle. You need top-left corner and bottom-right corner of rectangle.

```python
image_new = image.copy()
cv2.rectangle(image_new,(384,0),(510,128),(0,255,0),100)
plt.imshow(image_new);
```

#### Arithmetic Operations

You can add two images by OpenCV function, cv.add(). Res = img1 + img2. Both images should be of same depth and type.

```python
image = cv2.imread('media2/S1.jpg', -1)
image1 = cv2.imread('media2/S2.jpg', -1)
```

```python
plt.imshow(image)
```

```python
plt.imshow(image1)
```

```python
Add_image = cv2.add(image, image1)
plt.imshow(Add_image)
```

We can also subtract both images based on pixel density, with cv.subtract(). Res = img1 - img2. Both images should also be of same depth and type.

```python
B = cv2.subtract(image, image1)
plt.imshow(B)
cv2.imwrite("image1.jpg", B)
```
