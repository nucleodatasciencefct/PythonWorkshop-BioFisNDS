```python
#Getting libraries

import cv2
from skimage.measure import compare_ssim as ssim
from matplotlib import pyplot as plt
```

```python
#Load the two input images
S1 = 'S1.jpg'
S2 = 'S3.jpg'
imageA = cv2.imread(S1, -1)
imageB = cv2.imread(S2, -1)
```

```python
#Convert the images to grayscale
grayA = cv2.cvtColor(imageA, cv2.COLOR_BGR2GRAY)
grayB = cv2.cvtColor(imageB, cv2.COLOR_BGR2GRAY)
```

```python
#Display
plt.imshow(grayA, cmap = 'gray')
```

```python
plt.imshow(grayB, cmap = 'gray')
```

```python
#Score 
(score, diff) = ssim(grayA, grayB, multichannel=True, full=True)
diff = (diff * 255).astype("uint8")
print("SSIM: {}".format(score))
```

```python
#Get Resulting image

diff = cv2.subtract(grayA, grayB)
```

```python
#Save and Display 

cv2.imwrite("result.jpg", diff)
result = cv2.imread('result.jpg', -1)
plt.imshow(result)
```
